```python
import numpy as np
import matplotlib.pyplot as plt
meanIncome = 60000
sigmaIncome = 8000
lambdaIncome = 6.25e-5
scale_E = 1 / lambdaIncome
IncomeOfX = np.linspace(meanIncome - (4 * sigmaIncome), meanIncome + (4 * sigmaIncome), 1000)
ExpensesOfX = np.linspace(0, 2 * scale_E, 1000)
```

```python
from scipy.stats import norm, expon
incomePdf = norm.pdf(IncomeOfX, meanIncome, sigmaIncome)
expensesPdf = expon.pdf(ExpensesOfX, scale=scale_E) 
```

```python
plt.figure()
plt.subplot(2, 1, 1)
plt.plot(IncomeOfX, incomePdf)
plt.title('Income -- Normal Distribution')
plt.xlabel('Income')
plt.ylabel('Density')
plt.show()
```

```python
plt.figure()
plt.subplot(2, 1, 2)
plt.plot(ExpensesOfX, expensesPdf)
plt.title('Expenses -- Exponential Distribution')
plt.xlabel('Expenses')
plt.ylabel('Density')
plt.show()
```

```python
def monte_carlo_simulation(trial):
    incomes = np.random.normal(meanIncome, sigmaIncome, trial)
    expenses = np.random.exponential(scale_E, trial)
    profits = incomes - expenses
    return profits
```

```python
listOfTrials = [500, 1000, 50000]
results = {}
import seaborn as sns
```

```python
for trial in listOfTrials:
    profits = monte_carlo_simulation(trial)
    meanOfProfit = np.mean(profits)
    varOfProfit = np.var(profits)
    lossOfProbability = np.mean(profits < 0)
    
    results[trial] = {
        'mean': meanOfProfit,
        'variance': varOfProfit,
        'loss_probability': lossOfProbability
    }
    
    plt.figure()
    sns.histplot(profits,  kde=True, bins=30, color='blue')
    plt.title(f'Histogram of Total Profit (N={trial})')
    plt.xlabel('Profit')
    plt.ylabel('Frequency')
    plt.show()
```
