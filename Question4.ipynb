```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
numberOfSimulations = 4000  
sampleSizes = [200, 300, 500, 10000]
numberOfVariables = 75
```

```python
plt.figure(figsize=(14, 10))
for a, sampleSize in enumerate(sampleSizes):
    sums = np.zeros(numberOfSimulations)
    for i in range(numberOfSimulations):
        randomVariables = np.random.uniform(0, 1, size=(sampleSize, numberOfVariables))
        sums[i] = np.sum(randomVariables)
    
    plt.subplot(2, 2, a + 1)
    sns.histplot(sums, kde=True, stat="density", bins=50, color="mediumblue")
    plt.title('Sum of '+ str(sampleSize) +'Uniform Random Variables')
    plt.xlabel('Sum')
    plt.ylabel('Density')
    mean = sampleSize * numberOfVariables * 0.5  
    sigma = np.sqrt(sampleSize * numberOfVariables / 12) 
    x = np.linspace(np.min(sums), np.max(sums), 1000)
    normal_density = (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean) / sigma) ** 2)
    plt.plot(x, normal_density, color='darkorange', linestyle='--')
```
