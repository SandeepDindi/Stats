```python
import numpy as np
import matplotlib.pyplot as plt
# Generate random values between 1 and 7, we expect a mean around 4
x = np.random.uniform(1, 7, 100000)
meanVal, expectedMean =  np.mean(x), (1 + 7)/2
meanOfRoots = np.mean(np.sqrt(x))
meanOfSquares = np.mean(x ** 2)
meanOfInverse = np.mean(1 / x)

print("Numerical Mean of x:", meanVal)
print("Expected Mean Value of x (theoretical):", expectedMean)
print("Expected Mean Value of sqrt(x):", meanOfRoots)
print("Expected Mean Value of x^2:", meanOfSquares)
print("Expected Mean Value of 1/x:", meanOfInverse)

plt.figure()

plt.subplot(3, 1, 1)
plt.hist(np.sqrt(x), bins=50, color='blue', edgecolor='black')
plt.title('Histogram of sqrt(x)')
plt.xlabel('sqrt(x)')
plt.ylabel('Frequency')

plt.subplot(3, 1, 2)
plt.hist(x**2, bins=50, color='green', edgecolor='black')
plt.title('Histogram of x^2')
plt.xlabel('x^2')
plt.ylabel('Frequency')

plt.subplot(3, 1, 3)
plt.hist(1/x, bins=50, color='red', edgecolor='black')
plt.title('Histogram of 1/x')
plt.xlabel('1/x')
plt.ylabel('Frequency')
```
